```python
import nltk
from nltk.corpus import treebank

#treebank tagged words
treebank_tag = treebank.tagged_sents()
#Training Size
size = int(len(treebank_tag) * 0.9)

#Treebank training set
treebank_train = treebank_tag[:size]
#Treebank test set
treebank_test = treebank_tag[size:]

#Default Tagger
t0 = nltk.DefaultTagger('NN')
#Most Frequent tagger
t1 = nltk.UnigramTagger(treebank_train, backoff= t0)
#Bigram tagger
t2 = nltk.BigramTagger(treebank_train, backoff=t1)

print("Tagger Accuracy: {:.4f}\n".format(t2.evaluate(treebank_test)))

#Reading Emma
emma = nltk.corpus.gutenberg.raw('austen-emma.txt')
#Tokenizing Emma Sentences
emma_tokens = nltk.sent_tokenize(emma)
#Tokenizing sentences into words
emma_tokens = [nltk.word_tokenize(s) for s in emma_tokens]

#Tagging Emma sentences
emma_tagged = nltk.pos_tag_sents(emma_tokens)
#Flattening the text
emma_tagged_flat = [w for sent in emma_tagged for w in sent]

#Creating frequency dictionary
emma_dict = nltk.FreqDist(t for w, t in emma_tagged_flat)

#Getting the number of words
emma_len = len(emma_tagged_flat)

for tag, count in emma_dict.most_common(len(emma_dict.items())):
    print("Tag: {:s}  \tFrequency: {:.4f}".format(tag, count / emma_len))
```
